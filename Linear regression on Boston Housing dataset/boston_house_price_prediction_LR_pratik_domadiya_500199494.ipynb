# Load the dataset and libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
data = load_boston()
print(data.DESCR)
```

```python
print(data.feature_names) # display feature names of the dataset
```

```python
data.target# display target column, in our case target column is "MEDV"
```

```python
# convert data set to pandas dataframe
data_df = pd.DataFrame(data.data,columns=data.feature_names)
data_df
```

```python
# Target column is not included in data so we need to add it first.
data_df['MEDV']=data.target
(data_df.head(10))
```

# Data Analysis

```python
# check for missing values
data_df.isnull().sum()
# there is no any missing values are present.
```

```python
# finding correlation between all the attributes
data_df.corr()
```

```python
data_df.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
```

# split the data for training the model

```python
X_data = data_df.iloc[:, 0:13].values
y_data = data_df.iloc[:, 13].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.15, random_state = 25)
```

# standardize the training data

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X_ = StandardScaler()
sc_y_ = StandardScaler()
X_train_scaled = sc_X_.fit_transform(X_train)
y_train_scaled = sc_y_.fit_transform(y_train.reshape(-1,1))
```

# Train Linear regression model and display all the error metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train_scaled,y_train_scaled)
print( "Linear Regression..........  metrics")
print("model Training score : ",model.score(X_train_scaled, y_train_scaled))# find the accuracy of training dataset
# y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test)
# Predicting MSE the Test set results
mse_linear = ((mean_squared_error(y_test, y_pred_test)))
print(" MSE    : ",mse_linear)

# Predicting RMSE the Test set results
rmse_linear = (np.sqrt(mean_squared_error(y_test, y_pred_test)))
print(" RMSE  : ",rmse_linear)

# Predicting RMSE the Test set results
mae_linear = (np.sqrt(mean_absolute_error(y_test, y_pred_test)))
print(" MAE   :",mae_linear)
```

# Train Lasso regression model and display all the error metrics


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import Lasso
model= Lasso()
model.fit(X_train_scaled,y_train_scaled)
print( "Lasso Regression..........  metrics")
# print("model Training score : ",model.score(X_train_scaled, y_train_scaled))# find the accuracy of training dataset
y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test)
# Predicting MSE the Test set results
mse_linear = ((mean_squared_error(y_test, y_pred_test)))
print(" MSE    : ",mse_linear)

# Predicting RMSE the Test set results
rmse_linear = (np.sqrt(mean_squared_error(y_test, y_pred_test)))
print(" RMSE  : ",rmse_linear)

# Predicting RMSE the Test set results
mae_linear = (np.sqrt(mean_absolute_error(y_test, y_pred_test)))
print(" MAE   :",mae_linear)
```

# Train RandomForestRegressor model and display all the error metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
model= RandomForestRegressor(n_estimators=3000, max_depth=20, random_state=1)
model.fit(X_train_scaled,y_train_scaled)
print( "RandomForest Regression..........  metrics")

y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test)
print("model Training score : ",model.score(X_train_scaled, y_train_scaled))# find the accuracy of training dataset

# Predicting MSE the Test set results
mse_linear = ((mean_squared_error(y_test, y_pred_test)))
print(" MSE    : ",mse_linear)

# Predicting RMSE the Test set results
rmse_linear = (np.sqrt(mean_squared_error(y_test, y_pred_test)))
print(" RMSE  : ",rmse_linear)

# Predicting RMSE the Test set results
mae_linear = (np.sqrt(mean_absolute_error(y_test, y_pred_test)))
print(" MAE   :",mae_linear)
```

# Train Support Vector Regression model and display metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.svm import SVR
model = SVR(kernel = 'rbf', gamma = 'scale')
model.fit(X_train_scaled, y_train_scaled.ravel())
print( "Support Vector Regression..........  metrics")

y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test)
print("model Training score : ",model.score(X_train_scaled, y_train_scaled))# find the accuracy of training dataset

# Predicting MSE the Test set results
mse_linear = ((mean_squared_error(y_test, y_pred_test)))
print(" MSE    : ",mse_linear)

# Predicting RMSE the Test set results
rmse_linear = (np.sqrt(mean_squared_error(y_test, y_pred_test)))
print(" RMSE  : ",rmse_linear)

# Predicting RMSE the Test set results
mae_linear = (np.sqrt(mean_absolute_error(y_test, y_pred_test)))
print(" MAE   :",mae_linear)
```
